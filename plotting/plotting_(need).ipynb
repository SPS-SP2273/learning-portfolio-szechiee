# A simple plot

## Letâ€™s look at some code.

````python
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashdot')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=1)
plt.legend(loc='upper center')
````

|                   **Code**                   |                                                                                **Meaning**                                                                                |
|:--------------------------------------------:|:-------------------------------------------------------------------------------------------------------------------------------------------------------------------------:|
| plt.plot(x,y,label='',color='',linestyle='') | x = list of x values y = list of y values color = color of line linestyle = style of line, supported values (-, --, -., :, None, ' ', '', solid, dashed, dashdot, dotted) |
|                plt.xlabel('')                |                                                                          What to name the x-axis                                                                          |
|                plt.ylabel('')                |                                                                          What to name the y-axis                                                                          |
|                 plt.title('')                |                                                                    What to name the title of the graph                                                                    |
|              plt.grid(alpha=.25)             |                                             The degree of shading of the grid at the back of the plot, supported values (0-1)                                             |
|              plt.legend(loc='')              |  Location of the legend, supported values (best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center)  |nter  |

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
````

````python
plt.plot(x, y_1, c='red', ls='dashed', lw=2)
````

```color = c```, ```linestyle = ls```, ```linewidth = lw```

### Plotting multiple graphs

**Doing it together**

````python
plt.plot(x, y_1, y_2)
````

**Doing it separately**

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y_2, label='Y values',
           color='red', linestyle='dashed')
````

Order of keyword arguments does not matter.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

## Yet another plot but with error bars

````python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

# Better with NumPy

````python
import numpy as np
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
````

## Adding mathematical functions

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # square across all elements
sin_x = np.sin(x)          # sin(x) all elements

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
````

Plot does not look good because sin(x) lies between -1 and 1, but x^2 does not have such bounds, hence we need to add another y-axis.

## We need another axis!

### Things to note

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis                                     # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
````

Two legends() calls, one for each axis. Plot still does not look good as we only have a few points.

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis                                     # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right') 
````

Using ```np.linspace()``` to fix the problem as it returns evenly spaced numbers over a specified interval

# 3 Saving to disc

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')
#Saving the figure
plt.savefig('simple-plot.png', dpi=150)
plt.savefig('simple-plot.jpeg',dpi=300)
````

### Adding horizonal line at 0 value to highlight trend

````python
data = np.loadtxt('Complete_TAVG_daily.txt',   #load data from txt
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8]) 
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
````

## Styles

Matplotlib offers some standard [style templates](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

````python
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt',   #load data from txt
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8]) 
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
````

```plt.style.use('')``` must be included right at the top

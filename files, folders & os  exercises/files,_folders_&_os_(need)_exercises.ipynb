# Task 1

```python
dct = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
```

```python
def convert(text):
    month = text[:3]
    month = dct[month]
    date = text[4:6]
    country = text[7:-4]
    return f'{month}-{date}-{country}.txt'
```

```python
convert('oct-08_singapore.txt')
```

# Task 2

```python
def rename_my_file(old_file_name):
    new_name = convert(old_file_name)
    return new_name
```

# Task 3

```python
lst = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
```

```python
newlist = []
for txtfile in lst:
    newfile = rename_my_file(txtfile)
    newlist.append(newfile)
    print(f'{txtfile} ----> {newfile}')
print(newlist)
```

# Task 4

```python
import glob
import os
full_list = glob.glob('os-collaboration-exercise-data/**/*.txt', recursive=True)
for index in range(len(full_list)):
    try:
        filename = full_list[index]
        txtfile = os.path.split(filename)[-1]
        path = os.path.split(filename)[0]
        edited = rename_my_file(txtfile)
        new_name = os.path.join(path,edited)
        os.rename(filename,new_name)
    except:
        print('File already exists; skipping creation.')
```
